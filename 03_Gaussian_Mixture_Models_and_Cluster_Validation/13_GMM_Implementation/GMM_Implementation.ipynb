```python
from sklearn import datasets, mixture
```

## Load dataset

```python
iris = datasets.load_iris()
X    = iris.data[:10]
#
print("type(iris) = ", type(iris))
print("type(X)    = ", type(X))
print("X.shape    = ", X.shape)
```

### Specity the parameters for the clustering

```python
gmm = mixture.GMM(n_components=4)
gmm.fit(X)
clustering = gmm.predict(X)
```

```python
print("clustering      = ",     clustering)
print("len(clustering) = ", len(clustering))
```

```python
import matplotlib.pyplot as plt
tmp_x = list(range(10))
plt.plot(tmp_x, X[:,0], 'ro-');
plt.plot(tmp_x, X[:,1], 'go-');
plt.plot(tmp_x, X[:,2], 'bo-');
plt.plot(tmp_x, X[:,3], 'ko-');
plt.grid();
plt.legend(['0','1','2','3']);
```
